## Task 1: Cleaning the data
Note: Cleaning the data is generally a continuous thing. Often times, you won't know exactly what you'll need to clean until you run into a problem with the analysis. This is simply an introduction to some tools you can use to clean the data. 

```
# Take out the 2019 data because incomplete

df = df.loc[~df['season'].str.contains('2019-20')]
df.tail()
```

```
# Null or NaN values

# .isna() and .isnull() are aliases of each other and essentially do the same thing
#df.isna().sum()
df.isnull().sum()

# Adding additional parameters
    #  - Using .any() says select rows/columns that have NaN's in any of them
    #  - Using .all() says select rows/columns that are completely NaN values
    #  - Axis key: 0 is column, 1 is row
#df.isna().any(axis=0) 
#df.isna().all(axis=1)
```

```
# Setting decimal points
df.player_weight = df.player_weight.round(decimals=3)
df.head()
```

```
# Check for duplicated rows
df.duplicated(subset=None, keep='first').sum()
# Subset allows you to select a certain row or column
# Keep only the 'first' instance of the duplicate
# .sum() or .count() to count the number of duplicates
```

```
# Creating new variable season_year so that it can be more easily graphed/analyzed over time, etc.
# Basically an 'if statement'

df.loc[df['season'] == '1996-97', 'season_year'] = 1996
df.loc[df['season'] == '1997-98', 'season_year'] = 1997
df.loc[df['season'] == '1998-99', 'season_year'] = 1998
df.loc[df['season'] == '1999-00', 'season_year'] = 1999
df.loc[df['season'] == '2000-01', 'season_year'] = 2000
df.loc[df['season'] == '2001-02', 'season_year'] = 2001
df.loc[df['season'] == '2002-03', 'season_year'] = 2002
df.loc[df['season'] == '2003-04', 'season_year'] = 2003
df.loc[df['season'] == '2004-05', 'season_year'] = 2004
df.loc[df['season'] == '2005-06', 'season_year'] = 2005
df.loc[df['season'] == '2006-07', 'season_year'] = 2006
df.loc[df['season'] == '2007-08', 'season_year'] = 2007
df.loc[df['season'] == '2008-09', 'season_year'] = 2008
df.loc[df['season'] == '2009-10', 'season_year'] = 2009
df.loc[df['season'] == '2010-11', 'season_year'] = 2010
df.loc[df['season'] == '2011-12', 'season_year'] = 2011
df.loc[df['season'] == '2012-13', 'season_year'] = 2012
df.loc[df['season'] == '2013-14', 'season_year'] = 2013
df.loc[df['season'] == '2014-15', 'season_year'] = 2014
df.loc[df['season'] == '2015-16', 'season_year'] = 2015
df.loc[df['season'] == '2016-17', 'season_year'] = 2016
df.loc[df['season'] == '2017-18', 'season_year'] = 2017
df.loc[df['season'] == '2018-19', 'season_year'] = 2018

df
```
